# 1. Removal of Constant Features

```python
from sklearn.feature_selection import VarianceThreshold
sel = VarianceThreshold(threshold=0)
sel.fit(df)
```

```python
sum(sel.get_support())
```

```python
print(len([
        x for x in df.columns
        if x not in df.columns[sel.get_support()]
    ]))

constant_features=[x for x in df.columns if x not in df.columns[sel.get_support()]]
```

So there are 34 constant feature columns with same value in all the data out of 377 columns. So in a single step we are able to identify 38 features with zero variance which does not provide any predictive information of output variable. So we can straight away delete these columns. In the below statement we delete all the 34 constant feature columns.

```python
constant_features
```

```python
X_train = sel.transform(df)
df.shape
```

# 2. Quasi - Constant Features

These are the feature columns where most of the data has the same value. So in this section we consider the threshold as 99% ie., a column is quasi-constant feature column if more than 99% of the data has same value. We exclude all the quasi constant features as part of the pre-processing step. This is an important step when we have data with thousands of features and in these cases there is a high chance of having lot of constant features and quasi-constant features. It is ideal to test and remove these as they will have almost zero predictability of output variabe (Y).

The quasi-constant features and once identified delete those columns from original data.

```python
df = pd.read_csv('train.csv')
```

# 4.Pearson Correlation:-

In statistics, the Pearson correlation coefficient, also referred to as Pearson's r, the Pearson product-moment correlation coefficient, or the bivariate correlation, is a statistic that measures linear correlation between two variables X and Y. It has a value between +1 and −1. 



```python
df = pd.read_csv('train.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df.head()
```

```python
df['TARGET']=le.fit_transform(df['TARGET'])
```

```python
col=df.columns
col=col.drop('TARGET')
```

```python
from scipy.stats import pearsonr
cor=[]
for data1 in col:
    corr, _ = pearsonr(df[data1], df['TARGET'])
    cor.append([data1,corr])
cor.sort(key=lambda x: x[1],reverse=True)
cor
```

It knows when it has reached the top of the hill in that taking another step does not increase the likelihood.
